```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 101)
```

```python
a = 1
b = 1
y = a*np.exp(-((x+2.5)**2)/b)
```

```python
plt.plot(x, y)
```

```python
sp = np.fft.fft(y)
freq = np.fft.fftfreq(x.shape[-1])
plt.plot(freq, sp.real**2, freq, sp.imag)
```
